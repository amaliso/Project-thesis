```python
import matplotlib.pyplot as plt
import numpy as np
```

Importing experimental data from specimen S100_1

```python
S100_1 = np.loadtxt("s100_1.csv", skiprows=1, delimiter=",")
S100_1_time, S100_1_load, S100_1_ext = S100_1[:,0], S100_1[:,1], S100_1[:,2]
```

Importing Abaqus force displacement curves for different material combinations

```python
abaqus1 = np.loadtxt("abaqus1.csv", skiprows=7, delimiter=",")
abaqus2 = np.loadtxt("abaqus2.csv", skiprows=7, delimiter=",")
abaqus3 = np.loadtxt("abaqus3.csv", skiprows=7, delimiter=",")
abaqus4 = np.loadtxt("abaqus4.csv", skiprows=7, delimiter=",")
abaqus5 = np.loadtxt("abaqus5.csv", skiprows=7, delimiter=",")
```

```python
abaqus1_ext, abaqus1_load = abaqus1[:,0], abaqus1[:,1]
abaqus2_ext, abaqus2_load = abaqus2[:,0], abaqus2[:,1]
abaqus3_ext, abaqus3_load = abaqus3[:,0], abaqus3[:,1]
abaqus4_ext, abaqus4_load = abaqus4[:,0], abaqus4[:,1]
abaqus5_ext, abaqus5_load = abaqus5[:,0], abaqus5[:,1]
```

```python
plt.plot(S100_1_ext-.8, S100_1_load-1, label="S100_1 Experiment")

plt.plot(abaqus1_ext, abaqus1_load, label = "Abq mat 1")
plt.plot(abaqus2_ext, abaqus2_load, label = "Abq mat 2")
plt.plot(abaqus5_ext, abaqus5_load, label = "Abq mat 3")
plt.plot(abaqus3_ext, abaqus3_load, label = "Abq mat 4")
plt.plot(abaqus4_ext, abaqus4_load, label = "Abq mat 5")


plt.xlabel("Displacement [mm]")
plt.ylabel("Force [N]")
plt.legend()
plt.grid()
```
