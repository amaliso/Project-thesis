```python
def linear_fit(x, a, b):
    return a * x + b
```

```python
def stiffness_estimation(force, displacement, start_perc, stop_perc):
    max_force = np.max(force)
    max_force_index = np.argmax(force) #index av max force 
    max_displacement = displacement[max_force_index]
    force_index = np.arange(max_force_index)

    force1 = force[force_index]
    displacement1 = displacement[force_index]

    selected_indices = np.where((force1 >= start_perc * force1.max()) & (force1 <= stop_perc * force1.max()))
    
    selected_displacement = displacement1[selected_indices]
    selected_force = force1[selected_indices]
    #max_displacement[max_force_index]
    
    # Fit the linear model to the selected data
    params, covariance = curve_fit(linear_fit, selected_displacement, selected_force)

    # Predicted values from the linear fit
    predicted_force = linear_fit(selected_displacement, *params)

    # Calculate R-squared value
    residuals = selected_force - predicted_force
    ss_residual = np.sum(residuals**2)
    ss_total = np.sum((selected_force - np.mean(selected_force))**2)
    r_squared = 1 - (ss_residual / ss_total)

    # Print the R-squared value
    # Extract the slope (stiffness) from the fit parameters
    stiffness1 = params[0]
    '''
    plt.plot(displacement, force,linestyle='-', label='Force- Displacement Curve')
    plt.plot(selected_displacement, linear_fit(selected_displacement, *params), linestyle='-', color='r', label='Linear Fit')
    plt.scatter(max_displacement, force1.max(), color='red', label='Maximum force')
    plt.xlabel("Displacement [mm]")
    plt.ylabel("Force [N]")
    plt.title("Force displacement-curve of sample S110_1")
    plt.grid()
    plt.show()
    # Display the stiffness value
    print("R-squared value:", r_squared)
    #print("Estimated Stiffness_curvefit:", stiffness1, "N/mm")
    '''
    return stiffness1, max_force
    
```

```python
def stiffness_plateu(force, displacement, start_perc, stop_perc, plateau_end):
    plateau_indices = np.where((displacement >= 0) & (displacement <= plateau_end))[0]
  
    plat_displacement = displacement[plateau_indices]
   
    plat_force = force[plateau_indices]
    
    max_force_index = np.argmax(force[plateau_indices])
    max_force = force[plateau_indices][max_force_index]
    max_displacement = displacement[plateau_indices][max_force_index]

    selected_indices = np.where((plat_force >= start_perc * max_force.max()) & (plat_force <= stop_perc * max_force.max()))
    selected_displacement = displacement[selected_indices]
    selected_force = force[selected_indices]


    # Fit the linear model to the selected data
    params, covariance = curve_fit(linear_fit, selected_displacement, selected_force)

    # Predicted values from the linear fit
    predicted_force = linear_fit(selected_displacement, *params)

    # Calculate R-squared value
    residuals = selected_force - predicted_force
    ss_residual = np.sum(residuals**2)
    ss_total = np.sum((selected_force - np.mean(selected_force))**2)
    r_squared = 1 - (ss_residual / ss_total)
   
    # Extract the slope (stiffness) from the fit parameters
    stiffness1 = params[0]
    '''
    plt.plot(displacement, force,linestyle='-')
    plt.plot(selected_displacement, linear_fit(selected_displacement, *params), linestyle='-', color='r', label='Linear Fit')
    plt.scatter(max_displacement, max_force, color='red', label='Max Plateau Value')
    plt.xlabel("Displacement [mm]")
    plt.ylabel("Force [N]")
    plt.grid()
    plt.title("Force displacement-curve of sample S70_1")
    plt.show()
    # Display the stiffness value
    #print("R-squared value:", r_squared)
    #print("Estimated Stiffness_curvefit:", stiffness1, "N/mm")
    '''
    return stiffness1, max_force
```

```python

def total_energy(force, displacement):
    tot_e = np.trapz(force, displacement)
    return tot_e

#for 75-110 cases
def energy_max_f(force, displacement):
    # Find the index corresponding to the maximum force
    max_force_index = np.argmax(force)

    energy_up_to_max_force = np.trapz(force[:max_force_index], displacement[:max_force_index])

    return energy_up_to_max_force

def energy_max_plat(force, displacement, max_displacement):
    plateau_indices = np.where((displacement >= 0) & (displacement <= max_displacement))[0]
  
    plat_displacement = displacement[plateau_indices]
    plat_force = force[plateau_indices]

    max_force_index = np.argmax(plat_force)

    energy_up_to_max_force = np.trapz(plat_force[:max_force_index], plat_displacement[:max_force_index])

    #plt.plot(displacement, force)
    #plt.plot(plat_displacement, plat_force)
    
    return energy_up_to_max_force
```
